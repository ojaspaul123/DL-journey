MLP memorization refers to the tendency of a Multi-Layer Perceptron (MLP) to remember training examples instead of learning general patterns that work on unseen data.

Intuition

Suppose you train an MLP to classify cats and dogs.

Good learning (generalization): Learns features like ears, eyes, fur patterns.
Memorization (overfitting): Learns that image #123 has a particular pixel arrangement and simply remembers it.

When memorization dominates, training accuracy becomes very high, but test accuracy drops.

**Code-Demo**

```python
def fib (n):
  if n==1 or n==0 :
    return 1
  else :
    return fib(n-1) + fib(n-2)
```

```python
fib(40)
```

```python
import time
def fib (n):
  if n==1 or n==0 :
    return 1
  else :
    return fib(n-1) + fib(n-2)
```

```python
st = time.time()
fib(40)
old_end = time.time()-st
print("Time req.: ",old_end)
```

**Memorization**

```python
import time
def fib (n,d):
  if n in d:
    return d[n]
  else :
    d[n] = fib(n-1,d) + fib(n-2,d)
    return d[n]
```

```python
st = time.time()
d = {0:1,1:1}
fib(40,d)
new_end = time.time()-st
print("Time req.: ",new_end)
```

```python
print(f"Using memorization in new program is {old_end - new_end}x faster than old program.")
```
