```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
or_data = pd.DataFrame()
```

```python
or_data['input1'] = [1,1,0,0]
or_data['input2'] = [1,0,1,0]
or_data['output'] = [1,1,1,0]
```

```python
or_data
```

```python
and_data = pd.DataFrame()
```

```python
and_data ['input1'] = [1,1,0,0]
and_data ['input2'] = [1,0,1,0]
and_data ['output'] = [1,0,0,0]
```

```python
and_data
```

```python
Xor_data = pd.DataFrame()
```

```python
Xor_data['input1'] = [1,1,0,0]
Xor_data['input2'] = [1,0,1,0]
Xor_data ['output'] = [0,1,1,0]
```

```python
Xor_data
```

```python
sns.scatterplot(x='input1',y='input2',hue='output',data=or_data,s = 100)
```

```python
sns.scatterplot(x='input1',y='input2',hue='output',data=and_data,s = 100)
```

```python
sns.scatterplot(x='input1',y='input2',hue='output',data=Xor_data,s = 100)
```

```python
from sklearn.linear_model import Perceptron
c1 = Perceptron()
c2 = Perceptron()
c3 = Perceptron()
```

```python
c1.fit(or_data[['input1','input2']],or_data['output'])
c2.fit(and_data[['input1','input2']],and_data['output'])
c3.fit(Xor_data[['input1','input2']],Xor_data['output'])
```

```python
c1.coef_
```

```python
c1.intercept_
```
