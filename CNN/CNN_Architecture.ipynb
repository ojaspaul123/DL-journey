LeNet-5, introduced by Yann LeCun in 1998, was one of the earliest convolutional neural networks and was originally designed to recognize handwritten digits (it powered early bank check and zip code readers). It's a relatively shallow network by today's standards, but it established the core CNN pattern that's still used now: alternating convolution and pooling layers to extract spatial features, followed by fully connected layers to do the final classification.   


The architecture flows through 7 layers (excluding the input):

* Input takes a 32x32 grayscale image — larger than the actual MNIST digit (28x28) because padding helps the convolution filters capture features even at the edges of the digit.
* C1 and C3 are convolutional layers that scan the image with small 5x5 filters, building up "feature maps" that detect things like edges and curves. Notice the number of feature maps grows from 6 to 16 — each layer learns more distinct features even as the spatial size shrinks.
* S2 and S4 are pooling layers that downsample the feature maps, making the network more tolerant to small shifts or distortions in the input.
* C5, F6, and Output are the fully connected layers that take the extracted features and map them down to a final decision — in the original use case, 10 output units for the digits 0 through 9.
