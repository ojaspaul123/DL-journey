```python
from keras.applications.vgg16 import VGG16
```

```python
model = VGG16()
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model)
```

```python
for i in range(len(model.layers)):
  if 'conv' not in model.layers[i].name:
    continue

  filters,biases = model.layers[i].get_weights()
  print("layer number",i,model.layers[i].name,filters.shape)
```

```python
filters,biases = model.layers[1].get_weights()
```

```python
f_min, f_max = filters.min(),filters.max()
filters = (filters - f_min)/ (f_max - f_min)
```

```python
import matplotlib.pyplot as pyplot
```

```python
n_filter = 6
ix =1
fig = pyplot.figure(figsize=(10,10))
for i in range(n_filter):
  f = filters[:,:,:,i]
  for j in range(3):
    ax = pyplot.subplot(n_filter,3,ix)
    pyplot.imshow(f[:,:,j],cmap='gray')
    ix += 1

pyplot.show()
```

```python
from tensorflow.keras.models import Model
model = Model(inputs= model.inputs,outputs=model.layers[1].output)
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.vgg16 import preprocess_input

image = load_img('/content/ms.jpg', target_size=(224,224))
image = img_to_array(image)
image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
image = preprocess_input(image)
```

```python
features = model.predict(image)

fig = pyplot.figure(figsize=(10,10))

for i in range(features.shape[3]):
    pyplot.subplot(8, 8, i+1)
    pyplot.imshow(features[0, :, :, i], cmap='gray')

pyplot.show()
```

```python
model2 = VGG16()
```

```python
layer_index = [2,5,9,13,17]

outputs = [model2.layers[i].output for i in layer_index]
model2 = Model(inputs=model2.inputs,outputs=outputs)
```

```python
import math

features_map = model2.predict(image)

for i, fmap in zip(layer_index, features_map):
    n_filters = fmap.shape[3]
    grid_size = math.ceil(math.sqrt(n_filters))

    fig = pyplot.figure(figsize=(10,10))
    fig.suptitle(model2.layers[i].name)
    for j in range(n_filters):
        pyplot.subplot(grid_size, grid_size, j+1)
        pyplot.axis('off')
        pyplot.imshow(fmap[0, :, :, j], cmap='gray')
    pyplot.show()
```
