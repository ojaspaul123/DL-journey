**Padding**

*Padding* means adding extra pixels (usually zeros) around the border of the input image before performing convolution.

Why use padding?
* Prevents the output from shrinking too quickly after multiple convolutions.
* Allows edge pixels to be processed as much as center pixels.
* Helps preserve spatial information.

**Example**

image.png

**Code Demo**

```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,Flatten
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

**Stride**

*Stride* is the number of pixels the filter moves after each convolution operation.

Why use stride?
* Controls the size of the output feature map.
* Larger strides reduce computation and memory usage.
* Acts like downsampling.

**Example**

image.png

image.png

**Code-Demo**

```python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```
